```python
# different hosts & area

host_areas = airbnb.groupby(['host.host_name', 'address.street'])['host.host_total_listings_count'].max().reset_index()
host_areas.sort_values(by= 'host.host_total_listings_count',ascending= False).head(10)
```

```python
# different types of rooms and its prices (Overall)

room_price = airbnb.groupby(['address.street','room_type'])['price'].max().reset_index()
room_price.sort_values(by= 'price', ascending= False).head(10)
```

```python
# different types of rooms and its prices (Country wise)

room_price_cw = airbnb.groupby(['address.country','room_type'])['price'].max().reset_index()
room_price_cw.sort_values(by= 'price', ascending= False).head(10)
```

```python
# Visualize for a room type according to address (Top10)

Address_Country = ['Turkey','Hong Kong','Hong Kong','Brazil','Hong Kong','Turkey','United States','Spain','Brazil','Australia']

Room_Type = ['Private room','Private room','Shared room','Entire home/apt','Entire home/apt','Entire home/apt','Entire home/apt','Entire home/apt','Private room','Entire home/apt']

room_type_dict = {}

for i in Room_Type:
    room_type_dict[i] = room_type_dict.get(i, 0) + 1

plt.bar(room_type_dict.keys(), room_type_dict.values(), color= 'blue')
plt.title('TYPES OF ROOM')
plt.xlabel('Room Type')
plt.ylabel('Count of each')
plt.show
```

```python
#Visualize for Price according to the places

price_area = airbnb.groupby(['price'])['address.street'].max().reset_index()
price_area.head(10)
```

```python
price_list = price_area['price']
place = price_area['address.street']
fig = plt.figure(figsize= (50,50))

plt.scatter(place, price_list)
plt.xlabel("Countries")
plt.ylabel("Price")
plt.xticks(place,rotation=45)
plt.show
```

```python
#visualize for price according to the countries

price_area = airbnb.groupby(['price'])['address.country'].max().reset_index()
price_area.head(10)
```

```python
price_list = price_area['price']
place = price_area['address.country']
# fig = plt.figure(figsize= (10,5))

plt.scatter(place, price_list)
plt.xlabel("Countries")
plt.ylabel("Price")
plt.title('Price According to Countries')
plt.xticks(place,rotation=90)
plt.show
```

```python
# Find the Most busiest host 

busy_host = airbnb.groupby(['host.host_id','host.host_name','room_type'])['host.host_total_listings_count'].max().reset_index()
busy_host = busy_host.sort_values(by= 'host.host_total_listings_count', ascending=False).head(10)
busy_host
```

```python
name_host = busy_host['host.host_name']
listings = busy_host['host.host_total_listings_count']

plt.bar(name_host, listings, color= 'green', width= 0.5)
plt.xlabel('Name of the Host')
plt.ylabel('Total Listings')
plt.title('Busiest Host (according to no of listings)')
plt.xticks(name_host,rotation=90)
plt.show
```

```python
# Find and Visualise the Expensive host

expensive_place = airbnb.groupby(['host.host_id','host.host_name','room_type','address.street', 'address.country'])['price'].max().reset_index()
expensive_place = expensive_place.sort_values(by= 'price', ascending=False).head(10)
expensive_place
```

```python
name_of_host = expensive_place['host.host_name']
price_for_place = expensive_place['price']

plt.bar(name_of_host, price_for_place, color= 'red', width= 0.5)
plt.xlabel('Name the Host')
plt.ylabel('Price')
plt.title('Expensive Host (according to Price)')
plt.xticks(name_of_host,rotation=90)
plt.show
```

```python
# Find the Type of place people wanted to stay 


like_to_stay = airbnb.groupby(['address.street', 'room_type'])['minimum_nights'].count().reset_index()
like_to_stay = like_to_stay.sort_values(by= ['minimum_nights'], ascending= False).head(10)
like_to_stay
```

```python
Type_of_Room = like_to_stay['room_type']
Days_Stay = like_to_stay['minimum_nights']

plt.bar(Type_of_Room, Days_Stay, color= 'purple')

plt.xlabel('Room Type')
plt.ylabel('Minimum Nights')
plt.title('Most Liked to Stay')
plt.show
```
